```python
import requests
from bs4 import BeautifulSoup
import json
import os
```

```python
MEMORY_FILE = "memory.json"
```

```python
def load_memory():
    if os.path.exists(MEMORY_FILE):
        with open(MEMORY_FILE, "r") as f:
            return json.load(f)
    return []
def save_memory(memory):
    with open(MEMORY_FILE, "w") as f:
        json.dump(memory, f, indent=2)
```

```python
def fetch_webpage(url):
    response = requests.get(url)
    response.raise_for_status()
    return response.text
def extract_headlines(html):
    soup = BeautifulSoup(html, "html.parser")
    headlines = []

    for item in soup.select(".titleline > a"):
        headlines.append(item.text.strip())

    return headlines[:5]  # Top 5 headlines
```

```python
class NewsAgent:
    def __init__(self):
        self.memory = load_memory()
        self.state = "IDLE"

    def run(self):
        print("Agent State:", self.state)

        self.state = "SCRAPING_NEWS"
        html = fetch_webpage("https://news.ycombinator.com")
        headlines = extract_headlines(html)
        print("Agent State:", self.state)

        self.state = "CHECKING_MEMORY"
        new_headlines = [h for h in headlines if h not in self.memory]
        print("Agent State:", self.state)

        if new_headlines:
            self.state = "UPDATING_MEMORY"
            self.memory.extend(new_headlines)
            save_memory(self.memory)
            print("Agent State:", self.state)

        self.state = "DONE"
        print("Agent State:", self.state)

        return new_headlines
```

```python
def send_email_simulation(headlines):
    print("\n DAILY NEWS EMAIL")
    print("-" * 30)
    for h in headlines:
        print("•", h)
```

```python
agent = NewsAgent()
new_news = agent.run()

if new_news:
    send_email_simulation(new_news)
else:
    print("\n No new news today.")
```
